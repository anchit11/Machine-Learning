# Intro to jupyter notebook

A Jupyter Notebook is a powerful and versatile open-source tool for interactive computing and data exploration. It provides an environment where users can seamlessly integrate live code, visualizations, and explanatory text in a single, shareable document. The notebook is organized into cells, each of which can contain code snippets, Markdown-formatted text, or visual outputs. This structure allows for an iterative and exploratory approach to programming and analysis, as users can run code cells individually and observe immediate results. Supporting various programming languages through different kernels, Jupyter Notebooks cater to a wide range of applications, from data science and machine learning to research and education. With features like rich output display, collaborative editing, and diverse export options, Jupyter Notebooks have become a popular choice for professionals and researchers seeking an interactive and collaborative computing environment.

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
    
print(factorial(9))
print(factorial(14))
```

```python
import ipywidgets as widgets
from IPython.display import display

def on_slider_change(change):
    print(f"Slider Value: {change['new']}")
    
slider = widgets.IntSlider(value=0, min=0, max=100, step=1, description='Slider')

slider.observe(on_slider_change, names='value')

display(slider)
```

# Jupyter notebook extensions

Jupyter Notebook is an open-source web application that allows users to create and share documents containing live code, equations, visualizations, and narrative text. It provides an interactive computing environment that supports various programming languages, including Python, R, and Julia. The key feature of Jupyter Notebook is the ability to create and execute code in a modular and visually appealing format. Users can create notebooks containing a mix of code cells and markdown cells, enabling the integration of code and explanatory text. This interactive and literate programming approach makes it well-suited for data analysis, research, and educational purposes. Jupyter Notebook promotes collaboration and reproducibility by allowing users to share their work as interactive documents, making it a popular tool in the fields of data science, research, and education.
